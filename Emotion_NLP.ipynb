```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores('There are moments when I wish I could roll back the clock and take all the sadness away, but I have the feeling that if I did, the joy would be gone as wel')
```

```python
sia.polarity_scores('This is the worst thing ever.')
```

```python
example
```

```python
sia.polarity_scores(example)
```

```python
# Run polarity score on the entire dataset
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
    text = row["Comment"]
    myid = i  # Use index as ID if 'Id' column is not available
    res[myid] = sia.polarity_scores(text)
```

```python
res
```

```python
# Convert results to DataFrame
sentiment_df = pd.DataFrame.from_dict(res, orient='index')
```

```python
df = df.reset_index(drop=True)  # Ensure index matches
```

```python
# Merge with the original dataset
df = pd.concat([df, sentiment_df], axis=1)
```

```python
import seaborn as sns
```

```python
# Plot the emotion distribution
plt.figure(figsize=(8, 5))
sns.countplot(x=df["Emotion"], palette="viridis", order=df["Emotion"].value_counts().index)
plt.xlabel("Emotion")
plt.ylabel("Count")
plt.title("Emotion Distribution in Comments")
plt.xticks(rotation=45)
plt.show()
```

```python
#Drop duplicate columns if they exist
sentiment_df = sentiment_df.loc[:, ~sentiment_df.columns.duplicated()]
```

```python
 # Visualize sentiment scores
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

sns.barplot(data=sentiment_df, x=sentiment_df.index, y='pos', ax=axs[0], color='green')
sns.barplot(data=sentiment_df, x=sentiment_df.index, y='neu', ax=axs[1], color='blue')
sns.barplot(data=sentiment_df, x=sentiment_df.index, y='neg', ax=axs[2], color='red')

axs[0].set_title('Positive Sentiment')
axs[1].set_title('Neutral Sentiment')
axs[2].set_title('Negative Sentiment')

for ax in axs:
    ax.set_xticklabels([])  # Hide x-axis labels to avoid clutter

plt.tight_layout()
plt.show()
```
